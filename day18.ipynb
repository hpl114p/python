### List comprehension

```python
stop_words = ["I", "love", "and", "to"]
input = "I love AI and listen to music"
```

```python
list_input = list(map(str, input.split()))
print(list_input)

list_input_update = [word for word in list_input if word not in stop_words]
print(list_input_update)
```

### Tuple

```python
import math
```

```python
my_tuple1 = (2, 3)
my_tuple2 = (3, 6)

# Câu 2:
def sum_vector(vector):
    sum = 0
    product = 1
    for i in range(len(vector)):
        sum += vector[i]
        product *= vector[i]
    return (sum, product)

print(sum_vector(my_tuple1))
print(sum_vector(my_tuple2))
        
```

```python
# Câu 3:
def distance(P, Q):
    sum_dis = 0
    for i in range(len(P)):
        sum_dis += pow((P[i] - Q[i]), 2)
    return math.sqrt(sum_dis)

print(distance(my_tuple1, my_tuple2))
```
